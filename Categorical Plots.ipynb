# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Bar Plot

Most basic categorical plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

input arguments: x = 'Categorical Column Name', y = 'Numerical Column Name', data = Data Frame

Gives us the average total bill for both categories

There's also a 4th input argument we can use: estimator = aggregate function

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

Now we can observe the standard deviation between the two categories

## Count Plot

same as bar plot, except estimator only counts, so we only need to input the X component

```python
sns.countplot(x='sex',data=tips)
```

## Box Plots and Violin Plots

Used to show dist of categorical data

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# We can make the boxplot a bit more fancy by adding the hue parameter
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# violin plot is similar to boxplot
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# We can split the violin plots between the two different hues
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Strip Plot
scatter plot where one variable is categorical

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# To make more sense of this, we can introduce the Jitter argument
# this adds random noise to allow for the data to be better understood
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# now we can add hue = 'sex' as well as split = True
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

## Swarm Plot
Comined idea of strip plot and violin plot, prevents overlapping points

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# We can also combine the swarm plot with the violin plot as well
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

## Factor Plot
most general form of all of these plots

sns.factorplot(x, y, data, kind) -- the 'kind' argument is the main defining factor contributing to the form of the factor plot

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```
