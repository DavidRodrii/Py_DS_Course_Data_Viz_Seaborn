# Style and Color

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
# Removing the spines at Top and Right
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
# Removing all spines
# Removing the spines at Top and Right
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

## Size and Aspect

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

## Scale and Context

Seaborn comes with the convenient set_context which allows us to override parameters and set scales. Lets say we want to use the plot above for a poster...we would then do

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

## Palettes and Colors

Available at http://matplotlib.org/examples/color/colormaps_reference.html


```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Spectral')
plt.tight_layout()
```
