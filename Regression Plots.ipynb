# Regression Plots: LM Plot

Allows us to display linear models with seaborn

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## LMPlot

LM - stands for 'linear model'

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
# this is alright, the only issue is that the aspect is too small to see properly
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=8)
```

aspect = ratio of height and width
